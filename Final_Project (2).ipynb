# Preprocessing

```python
df.head()

logging.info('Started preprocessing.')
logging.info('Editing necessary columns values.')
logging.info('Droping NULL values.')
```

```python
df['education'].unique()
```

```python
def column_edit_education(studied):
    
    if (studied==' 11th') | (studied==' 9th') | (studied==' 7th-8th') | (studied==' 5th-6th') | (studied==' 10th') | (studied==' 1st-4th') | (studied==' Preschool') | (studied==' 12th'):
        return 'Lower'
    else:
        return studied
```

```python
df['education']=df['education'].apply(column_edit_education)
```

```python
df['education'].unique()
```

```python
df['marital-status'].unique()
```

```python
def column_edit_marital_status(status):
    if (status==' Never-married') | (status==' Divorced') | (status==' Separated') | (status==' Widowed') :
        return 'not-married'
    else:
        return 'married'
```

```python
df['marital-status']=df['marital-status'].apply(column_edit_marital_status)
```

```python
df['marital-status'].unique()
```

```python
df['hours-per-week'].dtype
```

```python
def column_edit_hrs_per_week(hrs):
    if (0 < hrs < 10):
      return '<10'
    elif (10 <= hrs <= 20):
      return 'between 10,20'
    elif (21 <= hrs <= 30):
      return 'between 21,30'
    elif (31 <= hrs <= 40):
      return 'between 31,40'
    elif (41 <= hrs <= 50):
      return 'between 41,50'
    elif (51 <= hrs <= 60):
      return 'between 51,60'
    elif (61 <= hrs <= 70):
      return 'between 61,70'
    elif (71 <= hrs <= 80):
      return 'between 71,80'
    elif (81 <= hrs <= 90):
      return 'between 81,90'
    elif (91 <= hrs <= 100):
      return 'between 91,100'
    else:
      return '>100'
```

```python
df['hours-per-week'] = df['hours-per-week'].apply(column_edit_hrs_per_week)
```

```python
def column_edit_salary(inc):
    if inc ==  ' <=50K':
        return 'Income less than 50K'
    else:
        return 'Income More then 50K'
```

```python
df['salary'] = df['salary'].apply(column_edit_salary)
```
